```python
#imports

import pandas as pd
import numpy as np
from selenium import webdriver
import datetime
from datetime import date
```

```python
#selenium driver configuration 

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
#scraping function

def crossword_data(html):
    #split crossword puzzle image into html elements + initialize matrix
    tag_splits = [i.split('>') for i in html.split('\n') if '<div class="num"' in i]
    cross_matrix = np.empty(0)
    #split text + get clue columns by ACROSS + DOWN, initialize clue direction, number, clue, and answer arrays
    split_txt = html.split('\n')
    clue_splits = [i.split('>') for i in [split_txt[i+1] for i in range(len(split_txt)) if \
                   '<div class="clueshead"' in split_txt[i]]]
    clue_dir = []
    clue_num = []
    clue = []
    answer = []

    #if the cell in each row is numbered, append a 1 to array, if it is a blackspace, append 0, reshape to appropriate crossword size
    for i in range(len(tag_splits)):
        row = []
        for j in tag_splits[i]:
            if j.replace('\t\t\t','') == '<div class="num"':
                row.append(1)
            elif j.replace('\t\t\t','') == '<td class="black"':
                row.append(0)
            else:
                pass
        cross_matrix = np.append(cross_matrix, [[row]])
        cross_matrix = cross_matrix.reshape(i+1, len(row))
    
    #split clue splits by specific html elements, append cleaned appropriate element to corresponding list
    #first half of clue splits is ACROSS, second half is DOWN
    for i in range(len(clue_splits)):
        for j in range(len(clue_splits[i])):
            try:
                clue_num.append(int(clue_splits[i][j].split('</')[0]))
                clue.append(clue_splits[i][j+2].split('<a')[0].rstrip(': '))
                answer.append(clue_splits[i][j+3].split('<')[0])
                if i == 0:
                    clue_dir.append('ACROSS')
                else:
                    clue_dir.append('DOWN')
            except:
                pass
            
    return cross_matrix, clue_dir, clue_num, clue, answer
```

```python
#today at this time was 11/3/2022
#initialize today string for pd date range and final data frame
#initialize webdriver + visit XWord site with login credentials
#initialize index for df

today = date.today()
today_str = str(today.timetuple()[0]) + '-' + str(today.timetuple()[1]) + '-' + str(today.timetuple()[2])
field_list = ['matrix', 'direction', 'clue_number', 'clue', 'answer', 'date']
final_df = pd.DataFrame(columns = field_list)

driver = webdriver.Chrome('chromedriver',options=chrome_options)
driver.get('https://www.xwordinfo.com/Account/Login.aspx')
driver.find_element("id", "UserName").send_keys('tleaves7')
driver.find_element("id", "Password").send_keys('pumpkin6')
driver.find_element("name", "ctl00$CPHContent$ctl02").click()
cnt = 0

#for each date in date range, visit the puzzle for the date, do the crossword scraping function, append date to output, append to dataframe
for i in pd.date_range(start="1993-11-21",end=today_str):
    try:
        curr = i.date()
        string_date = str(curr.timetuple()[1]) + '/' + str(curr.timetuple()[2]) + '/' + str(curr.timetuple()[0])
        print(string_date)
        driver.get(f'https://www.xwordinfo.com/Crossword?date={string_date}')
        html = driver.page_source
        temp = [i for i in crossword_data(html)]
        temp.append(string_date)
        final_df.loc[cnt] = temp
        cnt += 1
    except:
        pass
```
